### Estadística Descriptiva Proof


1. Demuestre $-1 \leq r_{XY}\leq 1$ donde:
$$
r_{XY}=\dfrac{\sum_{i=1}^n(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum_{i=1}^n(x_i-\bar{x})^2\sum_{i=1}^n(y_i-\bar{y})^2}}
$$

2. Demuestre que $|r_{XY} | = 1$ si y solo si existen una relación lineal entre las variables X y Y , es decir, $y_i = \alpha + \beta x_i  $ con $\beta \neq  0$

1. Sabemos que $r_{XY}$ es lo mismo que 

$$
r_{XY}:=\frac{\operatorname{Cov}(X, Y)}{\sigma_{X} \sigma_{Y}}
$$
donde $\sigma_{Z}^{2}$ es la varianza $V(Z)$ de una variable aleatoria $Z,$ y $\operatorname{Cov}(X, Y)$ es la covarianza definida como
$$
\operatorname{Cov}(X, Y):=\mathbb{E}\left(\left(X-\mu_{X}\right)\left(Y-\mu_{Y}\right)\right)=\mathbb{E}(X Y)-\mu_{X} \mu_{Y}
$$


Supongamos que $t$ es un número Real y que se cumple la siguiente desigualdad:

$$
\mathbb{E}\left((V+t W)^{2}\right) \geq 0, \text { donde } V=X-\mu_{X} \text { y } W=Y-\mu_{Y}
$$

Si expandemos del lado izquierdo y usamos la linearidad de la esperanza tenemos que

$$
\mathbb{E}\left(V^{2}\right)+2 t \mathbb{E}(V W)+t^{2} \mathbb{E}\left(W^{2}\right) \geq 0
$$

Podemos ver que el lado izquierdo es un polinomio cuadratico en $t$. Entonces tenemos que

$$
\mathbb{E}\left(V^{2}\right)=\sigma_{X}^{2}, \mathbb{E}\left(W^{2}\right)=\sigma_{Y}^{2}, \text { and } \mathbb{E}(V W)=\operatorname{Cov}(X, Y)
$$

Entonces nuestra desigualdad polinomica nos queda como

$$
\sigma_{Y}^{2} t^{2}+2 \operatorname{Cov}(X, Y) t+\sigma_{X}^{2} \geq 0
$$

De esta desigualdad podemos ver que la unica manera de que el lado izquierdo sea 0 es si el polinomio tiene  una raiz doble (es decir,toca al eje $x$ en un solo punto), lo cual ocurre solo si el discriminante es $0 .$ Entonces el discriminante siempre tiene que ser negativo o $0,$ lo que significa que 

$$
4 \operatorname{Cov}(X, Y)^{2}-4 \sigma_{X}^{2} \sigma_{Y}^{2} \leq 0
$$

Es decir,

$$
\frac{\operatorname{Cov}(X, Y)^{2}}{\sigma_{X}^{2} \sigma_{Y}^{2}} \leq 1
$$

2. De la prueba anterior podemos ver que la unica manera de que  $r_{XY}=\pm 1$ es si el discriminante del polinomio cuadratico es $0$, lo que significaria que el polinomio cuadratico se anulara para algun $t_{0}$. Esto significaria que:

$$
\mathbb{E}\left(\left(Y-\mu_{Y}+t_{0} X-t_{0} \mu_{X}\right)^{2}\right)=\mathbb{E}\left(\left(V+t_{0} W\right)^{2}\right)=0
$$

Esto solo ocurre si  $Y-\mu_{Y}+t_{0} X-t_{0} \mu_{X}=0$ con probabilidad 1, esto muestra que $X$ y $Y$ estan relacionados linealmente con probabilidad 1.

Ahora supongamos que

$$
Y=\lambda_{1} X+\lambda_{2}
$$

Entonces tenemos que $\mu_{Y}=\lambda_{1} \mu_{X}+\lambda_{2} ;$ y por tanto,

$$
\operatorname{Cov}(X, Y)=\mathbb{E}\left(\left(X-\mu_{X}\right)\left(\lambda_{1} X-\lambda_{1} \mu_{X}\right)\right)=\lambda_{1} \mathbb{E}\left(\left(X-\mu_{X}\right)^{2}\right)=\lambda_{1} \sigma_{X}^{2}
$$

Y por las propiedades de la varianza,

$$
\sigma_{Y}^{2}=V\left(\lambda_{1} X+\lambda_{2}\right)=V\left(\lambda_{1} X\right)=\lambda_{1}^{2} \sigma_{X}^{2}
$$

Lo que implica que

$$
r_{XY}=\frac{\operatorname{Cov}(X, Y)}{\sigma_{X} \sigma_{Y}}=\frac{\lambda_{1}}{\left|\lambda_{1}\right|}
$$

que es $\pm 1,$ y por lo tanto

$$
|r_{XY} | = 1
$$
