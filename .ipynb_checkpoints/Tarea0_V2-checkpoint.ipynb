####  Complejidad Computacional Notación O(n) (asymptotic upper bound)

Demuestre lo siguiente

1. Si $f_1$ es $O(g_1)$ y $f_2$ es $O(g_2)$ entonces $(f_1+f_2)$ es $O(max{|g_1|,|g_2|})$
2. Si $f_1$ es $O(g_1)$ y $f_2$ es $O(g_2)$ entonces $(f_1f_2)$ es $O(g_1g_2)$
3. ¿Cuál es la complejidad del siguiente algoritmo(O)?

for (i = 0; i < N; i++) {

    for (j = 0; j < M; j++) {
        print(i,j)
    }

}


**1.  Demostración:** 

Como $f_1$ es $O(g_1)$ entonces existen constantes positivas $c_1$ y $n_1$ tales que: $0 \le f_1(n) \le c_1g_1(n)$ $\forall n\ge n_1$

De la misma manera, como  $f_2$ es $O(g_2)$ entonces existen constantes positivas $c_2$ y $n_2$ tales que: $0 \le f_2(n) \le c_2g_2(n)$ $\forall n\ge n_2$

Sean:
$$n_0 = max \left\lbrace n_1, n_2 \right\rbrace$$
$$c = max \left\lbrace c_1, c_2 \right\rbrace$$

Se cumple que: $0 \le f_1(n) \le c_1g_1(n)$ $\forall n\ge n_0$ y $0 \le f_2(n) \le c_2g_2(n)$ $\forall n\ge n_0$

Sumando las dos desigualdades anteriores se tiene que: 
$$0 \le f_1(n) + f_2(n) \le c_1g_1(n) + c_2g_2(n) \forall n\ge n_0$$
$$\le c_1 max \left\lbrace |g_1(n)|, |g_2(n)| \right\rbrace + c_2 max \left\lbrace |g_1(n)|, |g_2(n)| \right\rbrace$$
$$\le c(max \left\lbrace |g_1(n)|, |g_2(n)| \right\rbrace)$$

Es decir, existen constantes positivas $c$ y $n_0$ tales que $0 \le f_1(n) + f_2(n) \le c(max \left\lbrace |g_1(n)|, |g_2(n)| \right\rbrace)$ $\forall  n\ge n_0$

$\therefore (f_1+f_2)$ es $O(max{|g_1|,|g_2|})$ 

**2.  Demostración:** 

Como $f_1$ es $O(g_1)$ entonces existen constantes positivas $c_1$ y $n_1$ tales que: $0 \le f_1(n) \le c_1g_1(n)$ $\forall n\ge n_1$

De la misma manera, como  $f_2$ es $O(g_2)$ entonces existen constantes positivas $c_2$ y $n_2$ tales que: $0 \le f_2(n) \le c_2g_2(n)$ $\forall n\ge n_2$

Sean:
$$n_0 = max \left\lbrace n_1, n_2 \right\rbrace$$
$$c = c_1 c_2 $$

Se cumple que: $0 \le f_1(n) \le c_1g_1(n)$ $\forall n\ge n_0$ y $0 \le f_2(n) \le c_2g_2(n)$ $\forall n\ge n_0$

Multiplicando  las dos desigualdades anteriores se tiene que: 
$$0 \le f_1(n) f_2(n) \le c_1g_1(n) c_2g_2(n) $$
$$=  c g_1(n) g_2(n)$$

Es decir, existen constantes positivas $c$ y $n_0$ tales que $0 \le f_1(n) f_2(n) \le c g_1(n) g_2(n)$ $\forall  n\ge n_0$

$\therefore (f_1f_2)$ es $O(g_1,g_2)$ 

### Estadística Descriptiva Prueba


1. Demuestre $-1 \leq r_{XY}\leq 1$ donde:
$$
r_{XY}=\dfrac{\sum_{i=1}^n(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum_{i=1}^n(x_i-\bar{x})^2\sum_{i=1}^n(y_i-\bar{y})^2}}
$$

2. Demuestre que $|r_{XY} | = 1$ si y solo si existen una relación lineal entre las variables X y Y , es decir, $y_i = \alpha + \beta x_i  $ con $\beta \neq  0$

1. Sabemos que $r_{XY}$ es lo mismo que 

$$
r_{XY}:=\frac{\operatorname{Cov}(X, Y)}{\sigma_{X} \sigma_{Y}}
$$
donde $\sigma_{Z}^{2}$ es la varianza $V(Z)$ de una variable aleatoria $Z,$ y $\operatorname{Cov}(X, Y)$ es la covarianza definida como
$$
\operatorname{Cov}(X, Y):=\mathbb{E}\left(\left(X-\mu_{X}\right)\left(Y-\mu_{Y}\right)\right)=\mathbb{E}(X Y)-\mu_{X} \mu_{Y}
$$


Supongamos que $t$ es un número Real y que se cumple la siguiente desigualdad:

$$
\mathbb{E}\left((V+t W)^{2}\right) \geq 0, \text { donde } V=X-\mu_{X} \text { y } W=Y-\mu_{Y}
$$

Si expandemos del lado izquierdo y usamos la linearidad de la esperanza tenemos que

$$
\mathbb{E}\left(V^{2}\right)+2 t \mathbb{E}(V W)+t^{2} \mathbb{E}\left(W^{2}\right) \geq 0
$$

Podemos ver que el lado izquierdo es un polinomio cuadratico en $t$. Entonces tenemos que

$$
\mathbb{E}\left(V^{2}\right)=\sigma_{X}^{2}, \mathbb{E}\left(W^{2}\right)=\sigma_{Y}^{2}, \text { and } \mathbb{E}(V W)=\operatorname{Cov}(X, Y)
$$

Entonces nuestra desigualdad polinomica nos queda como

$$
\sigma_{Y}^{2} t^{2}+2 \operatorname{Cov}(X, Y) t+\sigma_{X}^{2} \geq 0
$$

De esta desigualdad podemos ver que la unica manera de que el lado izquierdo sea 0 es si el polinomio tiene  una raiz doble (es decir,toca al eje $x$ en un solo punto), lo cual ocurre solo si el discriminante es $0 .$ Entonces el discriminante siempre tiene que ser negativo o $0,$ lo que significa que 

$$
4 \operatorname{Cov}(X, Y)^{2}-4 \sigma_{X}^{2} \sigma_{Y}^{2} \leq 0
$$

Es decir,

$$
\frac{\operatorname{Cov}(X, Y)^{2}}{\sigma_{X}^{2} \sigma_{Y}^{2}} \leq 1
$$

2. De la prueba anterior podemos ver que la unica manera de que  $r_{XY}=\pm 1$ es si el discriminante del polinomio cuadratico es $0$, lo que significaria que el polinomio cuadratico se anulara para algun $t_{0}$. Esto significaria que:

$$
\mathbb{E}\left(\left(Y-\mu_{Y}+t_{0} X-t_{0} \mu_{X}\right)^{2}\right)=\mathbb{E}\left(\left(V+t_{0} W\right)^{2}\right)=0
$$

Esto solo ocurre si  $Y-\mu_{Y}+t_{0} X-t_{0} \mu_{X}=0$ con probabilidad 1, esto muestra que $X$ y $Y$ estan relacionados linealmente con probabilidad 1.

Ahora supongamos que

$$
Y=\lambda_{1} X+\lambda_{2}
$$

Entonces tenemos que $\mu_{Y}=\lambda_{1} \mu_{X}+\lambda_{2} ;$ y por tanto,

$$
\operatorname{Cov}(X, Y)=\mathbb{E}\left(\left(X-\mu_{X}\right)\left(\lambda_{1} X-\lambda_{1} \mu_{X}\right)\right)=\lambda_{1} \mathbb{E}\left(\left(X-\mu_{X}\right)^{2}\right)=\lambda_{1} \sigma_{X}^{2}
$$

Y por las propiedades de la varianza,

$$
\sigma_{Y}^{2}=V\left(\lambda_{1} X+\lambda_{2}\right)=V\left(\lambda_{1} X\right)=\lambda_{1}^{2} \sigma_{X}^{2}
$$

Lo que implica que

$$
r_{XY}=\frac{\operatorname{Cov}(X, Y)}{\sigma_{X} \sigma_{Y}}=\frac{\lambda_{1}}{\left|\lambda_{1}\right|}
$$

que es $\pm 1,$ y por lo tanto

$$
|r_{XY} | = 1
$$

Los 40 alumnos de una clase han obtenido las siguientes puntuaciones, sobre 50, en un
examen de Física.

```python
califfisica=[48, 47, 44, 42, 41, 39, 39, 38, 38, 38, 37, 36, 36, 35, 35, 34, 34, 34, 33,
32, 32, 31, 29, 28, 28, 27, 26, 25, 24, 23, 22, 20, 17, 15, 15, 13, 13, 11, 7, 3]
```

1. Calcular las medidas de tendencia central y de dispersi ́on, ademas encuentre el Cuartil 1 y 3 así como los Deciles 1 y 9. Elabore un gráfico de caja y brazos, así como el gráfico de distribución empírica.
2. ¿ Los datos tiene una distribuci ́on platicurtica?

Medidas de tendencia central*

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
import statistics
C_califfisica= np.array(califfisica)
media=np.mean(C_califfisica)
mediana=np.median(C_califfisica)
moda=statistics.mode(C_califfisica)
#print(C_califfisica)
print(pd.Index(C_califfisica).value_counts())
print("La media es: "+str(media))
print("La mediana es: "+str(mediana))
print("La moda es: "+str(moda))
```

## Medidas  de dispersion

```python
var=np.var(C_califfisica)
std=np.std(C_califfisica)
RIC=np.quantile(C_califfisica,.75)-np.quantile(C_califfisica,.25)
cv=(std/media)*100
q_75=np.quantile(C_califfisica,.75)
q_25=np.quantile(C_califfisica,.25)
d_10=np.percentile(C_califfisica, 10)
d_90=np.percentile(C_califfisica,90)
print("el 90 percentil es:"+str(d_90))
print("el 10 percentil es:"+str(d_10))
print("El 1er cuartil  es: "+str(q_25))
print("El 3er cuartil  es: "+str(q_75))
print("La varianza es: "+str(var))
print("La desviacion estandar es: "+str(std))
print("El rango intercuartil es: "+str(RIC))
print("El coeficiente de variacion: "+str(var))
```

## Grafica de caja

```python
plt.boxplot(C_califfisica)
plt.xlabel(" Diagrama de caja")
plt.ylabel("Calificaciones")
plt.show()
```

## Distribucion empirica

```python
def ecdf(C_califfisica):
    #n numero de datos
    n= len(C_califfisica)
    
    # ordenar a x
    x=np.sort(C_califfisica)
    
    #datos de y
    y=np.arange(1,n+1)/n
    
    return x,y
#calcular ecdf
x_1 , y_1 = ecdf(C_califfisica)

plt.plot(x_1 , y_1 ,marker='',linestyle ='solid')
plt.title("Distribucion empirica")
plt.show()
```

## Distribucion platicurtica

*Utilicemos la prueba de la curtosis*

```python
kurtosis(C_califfisica,fisher=True)
```

#### Como el resultado es menor a 0 se concluye, por el coeficiente de curtosis de Fisher, que es platicurtica.
