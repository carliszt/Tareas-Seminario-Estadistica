```python
califfisica=[48, 47, 44, 42, 41, 39, 39, 38, 38, 38, 37, 36, 36, 35, 35, 34, 34, 34, 33,
32, 32, 31, 29, 28, 28, 27, 26, 25, 24, 23, 22, 20, 17, 15, 15, 13, 13, 11, 7, 3]
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import skew
import statistics
C_califfisica= np.array(califfisica)
media=np.mean(C_califfisica)
mediana=np.median(C_califfisica)
moda=statistics.mode(C_califfisica)
print(C_califfisica)
print(pd.Index(C_califfisica).value_counts())
print("La media es: "+str(media))
print("La mediana es: "+str(mediana))
print("La moda es: "+str(moda))
var=np.var(C_califfisica)
std=np.std(C_califfisica)
RIC=np.quantile(C_califfisica,.75)-np.quantile(C_califfisica,.25)
cv=(std/media)*100
q_75=np.quantile(C_califfisica,.75)
q_25=np.quantile(C_califfisica,.25)
d_10=np.percentile(C_califfisica, 10)
d_90=np.percentile(C_califfisica,90)
print("el 90 percentil es"+str(d_90))
print("el 10 percentil es :"+str(d_10))
print("El 1er cuartil es: "+str(q_25))
print("El 3er cuartil es: "+str(q_75))
print("La varianza es: "+str(var))
print("La desviacion estandar es: "+str(std))
print("El rango intercuartil es: "+str(RIC))
print("El coeficiente de variacion: "+str(var))
plt.boxplot(C_califfisica)
plt.xlabel(" diagrama de caja")
plt.ylabel("Calificaciones")
plt.show()
#idea de distribucion empirica
def ecdf(C_califfisica):
    #n numero de datos
    n= len(C_califfisica)
    
    # ordenar a x
    x=np.sort(C_califfisica)
    
    #datos de y
    y=np.arange(1,n+1)/n
    
    return x,y
#calcular ecdf
x_1 , y_1 = ecdf(C_califfisica)

plt.plot(x_1 , y_1 ,marker='',linestyle ='solid')
plt.title("distribucion empirica")
plt.show()
#checamos si es platicurtica
skew(C_califfisica)

    
```
