#### *i)* Si tuviese que eliminar dos variables cuales serían y porqué. Use SelectKBest

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("covid.csv",header=0)

data=data.drop(["Unnamed: 0"],axis=1)
data
```

Notemos que tenemos a priori 10 covariables. Efecteumos un Label Encoder para transformar las variables categóricas en numéricas.


```python
#Apliquemos un StringIndexer
col=["SEXO","OBESIDAD","DIABETES","EPOC","ASMA","CARDIOVASCULAR","RENAL CRONICA","INMUNOSUPRESION","HIPERTENSION","TIPO PACIENTE"]
def stringIndexer(df,col):
    X=df
    for i in col:
        X[i+"strindexer"]=LabelEncoder().fit_transform(X[i])  # Pasando de string a numericos
        X=X.drop(i,1)
    return X
```

```python
X=stringIndexer(data, col)
y=X.pop("TIPO PACIENTEstrindexer")
X
```

```python
y
```

Ahora, utilicemos la función SelectKBest para seleccionar a 8 covariables top del conjuto de 10 respecto a los scores mas altos.

```python
from sklearn.feature_selection import SelectKBest, f_classif
fselect = SelectKBest(f_classif, k=8)  
scores_fit = fselect.fit(X, y)
features_score = scores_fit.scores_.round(2)

features_score
```

En una tabla, los resultados son los siguientes:

```python
features= list(X.columns.values)
selected_features = fselect.get_support([scores_fit])
feat_list = [ ]
for i in selected_features:
    feat_list.append({'Feature':features[i], 'Score': features_score[i]  })
features_select = pd.DataFrame(feat_list)
features_select.sort_values('Score',ascending =False)
```

```python
X.columns.values
```

Por lo tanto, dado que en la tabla anterior se mostraron las 8 covariables top respecto a su desempeño en el score. De esta manera, las dos variables a eliminar serían *ASMA* e *INMUNOSUPRESION*.
