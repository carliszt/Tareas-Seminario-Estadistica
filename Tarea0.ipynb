*1) Solucion:*

Medidas de tendencia central

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import skew
import statistics 
a_calif=np.array(calif)
media=a_calif.mean()
mediana=np.median(a_calif)
print(pd.Index(a_calif).value_counts())
moda=statistics.mode(a_calif)

```

```python
print("La media es: "+str(media))
print("La mediana es: "+str(mediana))
print("La moda es: "+str(moda))
```

Medidas de dispersion

```python
var=np.var(a_calif)
std=np.std(a_calif)
RIC=np.quantile(a_calif,.75)-np.quantile(a_calif,.25)
cv=(std/media)*100
q_75=np.quantile(a_calif,.75)
q_25=np.quantile(a_calif,.25)

print("La varianza es: "+str(var))
print("La desviacion estandar es: "+str(std))
print("El rango intercuartil es: "+str(RIC))
print("El coeficiente de variacion: "+str(var))
print("El 1er cuartil es: "+str(q_25))
print("El 3er cuartil es: "+str(q_75))

```

Elaboremos el diagrama de caja y brazos

```python
plt.boxplot(a_calif)
plt.xlabel(" Boxplot")
plt.ylabel("Calificaciones")
plt.show()
```

```python
plt.hist(a_calif,  bins=11,color="darkblue", edgecolor="black")

plt.xlabel("Calificaciones")
plt.ylabel("Frecuencia")
plt.show()
```
